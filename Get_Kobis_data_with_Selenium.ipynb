# Selenium으로 Kobis Data 가져오기  

```python
# 데이터 처리
import pandas as pd

# Crawling
import time
from bs4 import BeautifulSoup
from selenium import webdriver


def get_table():
    time.sleep(2)
    
    # 페이지의 html을 받아옴
    html = driver.page_source

    # dataframe으로 만들기
    result = pd.read_html(html)[0]
    
    return result


def get_data():
    while True:
        try:
            tb = get_table()
            for i in range(2, 10):
                driver.find_element_by_xpath(f'//*[@id="pagingForm"]/div/ul/li[{i}]/a').click()
                temp = get_table()
                tb = pd.concat([tb, temp], axis = 0)
            driver.find_element_by_xpath('//*[@id="pagingForm"]/div/a[3]').click()

        except:
            break
            
    # 인덱스 재정렬
    tb.reset_index(drop = True, inplace = True) 
    
    # 창 닫기
    driver.close()
    
    return tb

# 영화 검색
def search_movie(name):
    # Kobis 영화페이지로 이동
    driver.get('http://www.kobis.or.kr/kobis/business/mast/mvie/searchMovieList.do')
        
    # 검색
    driver.find_element_by_xpath('//*[@id="searchForm"]/div[1]/div[1]/div/input').send_keys(name)
    driver.find_element_by_xpath('//*[@id="searchForm"]/div[1]/div[5]/button[1]').click()
  
    # 검색결과 가져오기
    table = get_data()

    return table
    
# 사람 검색
def search_people(name, movie):    
    # Kobis 영화인페이지로 이동
    driver.get('http://www.kobis.or.kr/kobis/business/mast/peop/searchPeopleList.do')
    
    # 검색
    driver.find_element_by_xpath('//*[@id="searchForm"]/div[1]/div[1]/div/input').send_keys(name)
    driver.find_element_by_xpath('//*[@id="searchForm"]/div[1]/div[2]/div/input').send_keys(movie)
    driver.find_element_by_xpath('//*[@id="searchForm"]/div[1]/div[3]/button[1]').click()

    # 검색결과 가져오기
    table = get_data()

    return table


def get_db(mp):
    if (mp == 'm') or (mp == 'movie'):
        driver.get('http://www.kobis.or.kr/kobis/business/mast/mvie/searchMovieList.do')
        
    elif (mp == 'p') or (mp == 'person') or (mp == 'people') :
        driver.get('http://www.kobis.or.kr/kobis/business/mast/peop/searchPeopleList.do')
    
    # 검색결과 가져오기
    table = get_data()
    
    return table
    
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
```
